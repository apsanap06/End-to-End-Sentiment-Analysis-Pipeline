```python
# Group complex emotions into 3 simple buckets
def simplify_emotion(emotion):
    emotion = str(emotion).strip().title()
    positives = ['Positive', 'Joy', 'Excitement', 'Happy', 'Happiness', 'Gratitude', 
                 'Love', 'Amusement', 'Enjoyment', 'Admiration', 'Affection', 'Proud', 
                 'Contentment', 'Serenity', 'Hope', 'Enchantment', 'Thrill']
    negatives = ['Negative', 'Anger', 'Fear', 'Sadness', 'Disgust', 'Disappointed', 
                 'Bitter', 'Jealousy', 'Resentment', 'Frustration', 'Boredom', 
                 'Anxiety', 'Despair', 'Grief', 'Heartbreak', 'Loneliness']
    
    if emotion in positives: return 2   # 2 = Positive
    elif emotion in negatives: return 0 # 0 = Negative
    else: return 1                      # 1 = Neutral

df['target'] = df['Sentiment'].apply(simplify_emotion)

# 3. Convert text to numbers 
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X_words = vectorizer.fit_transform(df['Text'].fillna('')).toarray()
y_sentiment = df['target']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_words, y_sentiment, test_size=0.2, random_state=42)
print("Task 1 Done: Data grouped into 3 emotions and text vectorized!\n")
```

# Level 2

```python
# --- Importing for level 2 ---
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

print("--- LEVEL 2: INTERMEDIATE MODELS ---")

# --- TASK 1: BINARY LOGISTIC REGRESSION ---
# Filter data to ONLY include Positive (2) and Negative (0)
binary_df = df[df['target'].isin([0, 2])].copy()
X_bin_words = vectorizer.fit_transform(binary_df['Text'].fillna('')).toarray()

# Map the targets to 0 and 1 for binary classification
y_bin = binary_df['target'].map({0: 0, 2: 1})
Xb_train, Xb_test, yb_train, yb_test = train_test_split(X_bin_words, y_bin, test_size=0.2, random_state=42)

log_model = LogisticRegression()
log_model.fit(Xb_train, yb_train)
print(f"Task 1 Done: Binary Logistic Regression Accuracy: {log_model.score(Xb_test, yb_test):.2f}\n")
```

```python
# --- TASK 2: DECISION TREE CLASSIFIER ---
tree_model = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_model.fit(X_train, y_train)
print(f"Task 2 Done: Decision Tree Accuracy: {tree_model.score(X_test, y_test):.2f}")
```

```python
# --- Importing for level 3 ---
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC


print("--- LEVEL 3: ADVANCED MODELS ---")

# --- TASK 1: RANDOM FOREST CLASSIFIER ---
forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
forest_model.fit(X_train, y_train)

print(f"Task 1 Done: Random Forest Accuracy: {forest_model.score(X_test, y_test):.2f}\n")
```
